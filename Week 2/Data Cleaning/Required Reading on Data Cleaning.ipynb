#### Checking for Missing Values

```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Creating a Data Frame Including NaN
dataset = pd.DataFrame(np.random.randn(5, 3),
index=['a', 'c', 'e', 'f', 'h'],columns=['stock1','stock2', 'stock3'])
dataset.rename(columns={"one":'stock1',"two":'stock2',"three":'stock3'}, inplace=True)
dataset = dataset.reindex(['a', 'b', 'c', 'd', 'e','f', 'g', 'h'])
print (dataset)
```

```python
# Creating a Matrix of Random Values
import numpy as np
np.random.randn(5, 3)
```

```python
# Checking Null Cases
print (dataset['stock1'].isnull())
```

#### Handling the Missing Values
 - Replacing NaN with a Scalar Value :

```python
#  replaces all NaN cases with 0 values
print (dataset)
dataset.fillna(0)
```

- You can fill NaN cases forward and backward

```python
# Filling In Missing Values Forward
print (dataset)
dataset.fillna(method='pad')
```

â€¢ You can drop the missing values.

```python
# Dropping All NaN Rows
print (dataset)
dataset.dropna()
```

â€¢ You can replace the missing (or generic) values.

```python
# Using the replace() Function
print (dataset)
dataset.replace(np.nan, 0 )
```

### Automatically Replacing Matched Cases with NaN

```python
import pandas as pd
sales = pd.read_csv("Sales.csv", nrows=7, na_values=["n.a.", "not avilable"])
mydata = sales.head(7)
mydata
```

```python
import pandas as pd
sales = pd.read_csv("Sales.csv", nrows=7, na_values=["n.a.", "not avilable", -1])
mydata = sales.head(7)
mydata
```

```python
sales = pd.read_csv("Sales.csv", na_values = {"SALES_BY_REGION": ["n.a.", "not avilabl"],"JANUARY": ["n.a.", "not avilable", -1],"FEBRUARY": ["n.a.", "not avilable", -1]})
sales.head(20)
```

```python
# Defining and Calling Functions for Data Cleaning

def CleanData_Sales(cell):
    if (cell=="n.a." or cell=="-1" or cell=="not avilable"):
        return 0
    return cell
def CleanData_REGION(cell):
    if (cell=="n.a." or cell=="-1" or cell=="not avilable"):
        return 'AbuDhabi'
    return cell
```

```python
sales = pd.read_csv("Sales.csv", nrows=7, converters={"SALES_BY_REGION": CleanData_REGION,"JANUARY": CleanData_Sales,"FEBRUARY": CleanData_Sales,"APRIL": CleanData_Sales,})
sales.head(20)
```
