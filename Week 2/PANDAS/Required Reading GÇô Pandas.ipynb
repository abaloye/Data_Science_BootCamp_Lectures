### Reading Data into Pandas Dataframe

```python
import pandas as pd
titanic_data = pd.read_csv("titanic_data.csv")
titanic_data.head()
```

### Filtering Rows

```python
# returns a series of True and False. True will be returned for indexes where the Pclass column has a value of 1.

titanic_pclass1= (titanic_data.Pclass == 1)
titanic_pclass1
```

```python
titanic_pclass1= (titanic_data.Pclass == 1)
titanic_pclass1_data = titanic_data[titanic_pclass1]
titanic_pclass1_data.head()
```

```python
titanic_pclass_data = titanic_data[titanic_data.Pclass == 1]
titanic_pclass_data.head()
```

```python
ages = [20,21,22]
age_dataset = titanic_data[titanic_data["Age" ].isin(ages)]
age_dataset.head()
```

```python
# The isin operator takes a list of values and returns only those rows where the column used
# for comparison contains values from the list passed to the isin operator as a
# parameter
ages = [20,21,22]
ageclass_dataset = titanic_data[titanic_data["Age"].isin(ages) & (titanic_data["Pclass"] == 1) ]
ageclass_dataset.head()
```

```python
# Filterind dataframe using l and (&) and or (|) operators.
ages = [20,21,22]
ageclass_dataset = titanic_data[titanic_data["Age"].isin(ages) & (titanic_data["Pclass"] == 1) ]
ageclass_dataset.head()
```

### Filtering Columns


```python
titanic_data_filter = titanic_data.filter(["Name" , "Sex" , "Age" ])
titanic_data_filter.head()
```

```python
#  you can also drop columns that you don’t want in the dataset using drop() method. 
titanic_data_filter = titanic_data.drop(["Name" , "Sex" , "Age"], axis = 1)
titanic_data_filter.head()
```

### Concatenating Dataframes

```python
titanic_pclass1_data = titanic_data[titanic_data.Pclass == 1]
print (titanic_pclass1_data.shape)

titanic_pclass2_data = titanic_data[titanic_data.Pclass == 2]
print (titanic_pclass2_data.shape)
```

```python
# One way to concatenate datasets horizontallyis through the append()
final_data = titanic_pclass1_data.append(titanic_pclass2_data, ignore_index=True)
print (final_data.shape)
```

```python
# Another way to concatenate datasets horizontally is through the concat()
final_data = pd.concat([titanic_pclass1_data, titanic_pclass2_data])
print (final_data.shape)
```

```python
# ignore_index attribute is to reset dataset indexes
df1 = final_data[:200]
print (df1.shape)
df2 = final_data[200:]
print (df2.shape)

final_data2 = pd.concat([df1, df2], axis = 1, ignore_index = True)
print (final_data2.shape)
```

### Sorting Dataframes

```python
#  sort_ values() method is used to  sort the Pandas dataframe
age_sorted_data = titanic_data.sort_values(by=['Age' ])
age_sorted_data.head()
```

```python
# sort by descending order
age_sorted_data = titanic_data.sort_values(by=['Age' ], ascending = False)
age_sorted_data.head()
```

```python
# You can also pass multiple columns to the by attribute of the sort_values() function.
age_sorted_data = titanic_data.sort_values(by=['Age' ,'Fare' ], ascending = False)
age_sorted_data.head()
```

### Apply Function


```python
# The apply() function is used to apply a function (eg. lambda expression) on multiple rows or on rows of a particular column. 

updated_class = titanic_data.Pclass.apply(lambda x : x + 2)
updated_class.head()
```

```python
# We apply a mult() function which multiplies the parameter passed to it by 2 and returns the resultant value.
def mult(x):
    return x * 2

updated_class = titanic_data.Pclass.apply(mult)
updated_class.head()
```

```python
import pandas as pd
titanic_data = pd.read_csv("titanic_data.csv" )
titanic_data.head()

# The crosstab() function is used to plot cross tabulation between two columns.
pd.crosstab(titanic_data.Pclass, titanic_data.Age, margins=True)
```
