```python
# and we will bring in some timing functionality too from the timeit module
import timeit 
#NB: To turn date format by 'pd.to_datetime(df['Timestamp'])'
# we perform vectorised operation through the Timestamp

# lets look at some census data from the US
df=pd.read_csv('c:\\Users\\Feranmi\\Downloads\\census.csv')
df.head()
```

```python
# the first of the pandas method used is called method chaining. 
# the general idea behind the method chaining is that every method on an object returns  a refrence to that object. 
# the beauty of this is that you can condence many different operation on a DataFrame. for instance into one line or 
# at least one statement of a code.
```

```python
# The set_index() function is used to set the DataFrame index using existing columns.
# the rename function is used to rename columns
(df.where(df['SUMLEV']==50)
    .dropna() #to drop null values
    .set_index(['STNAME', 'CTYNAME'])
    .rename(columns={'ESTIMATEBASE2010':'Estimate Base 2010'})) 
```

```python
# Here is a more traditional more pandorable way, of writing this. there is nothing wrong with this coding in the functional sense, 
#you might even be able to understand it better as a new person to the language. it's just not as pandorable as the first example.

# First create a new DataFrame from the original

# use the overloaded indexing operator [] which drops nans
df=df[df['SUMLEV']==50]

# update te DataFrame to have a new index, we use inplace=True to do this inplace
df.set_index(['STNAME', 'CTYNAME'], inplace=True)

# set the column names
df.rename(columns={'ESTIMATEBASE2010':'Estimate Base 2010'})
```

```python
# now the key with any good idiom is to understand when it isn't helping you.
# in this case you can actually time both method and see which one runs faster
# we can the approach into a function and pass the function into the timeit 
# function to count the time the parameter number allows us to choose ho many 
#times we want to run the function. Here we will just set it to 10

# lets write a wrapper for our first function
def first_approach():
    global df
    
    return(df.where(df['SUMLEV']==50)
              .dropna()
              .set_index(['STNAME', 'CTYNAME'])
              .rename(columns={'ESTIMATEBASE2010':'Estimate Base 2010'}))
# read in our new dataset
df = pd.read_csv('c:\\Users\\Feranmi\\Downloads\\census.csv')

timeit.timeit(first_approach, number=10)
```

```python
# now lets test the second approach
# as you may notice we use our global variable df in the function. However changing a golbal variable inside a function will modify the variable even in a global scope 
# and we do not want that to happen in this case. Therefore, for selecting summary levels of only 50, i create a new ataFrame for those records

# lets run this for once and see how fast it is.
def second_approach():
    global df
    new_df=df[df['SUMLEV']==50]
    new_df.set_index(['STNAME', 'CTYNAME'], inplace=True)
    return new_df.rename(columns={'ESTIMATEBASE2010':'Estimate Base 2010'})
           
 #READ IN DATASET ANEW
df = pd.read_csv('c:\\Users\\Feranmi\\Downloads\\census.csv')

timeit.timeit(second_approach, number=10)
```

```python
# lets take a look at our census dataframe, we have five columns for population estimates with each columns corresponding with one year of the estimates. 
# its quite resonable to want to create some new columns for minimum and maximum value, the apply function is an easy way to do that.

# first we write a function that takes in a particular row of data, find the minimum and maximum value and return a new row of data.
def min_max(row):
    data=row[['POPESTIMATE2010',
              'POPESTIMATE2011',
              'POPESTIMATE2012',
              'POPESTIMATE2013',
              'POPESTIMATE2014',
              'POPESTIMATE2015']]
    return pd.Series({'min':np.min(data), 'max':np.max(data)})
```

```python
# then we just need to call apply on on the DataFrame
df.apply(min_max, axis= 'columns').head()

#note that columns in axis can be replaced by 0 and 1
```

```python
# here is an example where we have revised version of the function min_max instead of returning a seperate series to display the min and max 
# we add two new columns in the original DataFrame to store min and max.
def min_max(row):
    data=row[['POPESTIMATE2010',
               'POPESTIMATE2011',
               'POPESTIMATE2012',
               'POPESTIMATE2013',
               'POPESTIMATE2014',
               'POPESTIMATE2015']]
    row['max']=np.max(data)
    row['min']=np.min(data)
    return row
df.apply(min_max, axis='columns')
```

```python
# Here you can imagine how you might chain several apply calls lambdas together to create a reasonable yet succinct data manupulation script .
#A lambda function is a small anonymous function.
#A lambda function can take any number of arguments, but can only have one expression.
# one online example how you might calculate the max of a columns using the apply function 

rows=['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']
df.apply(lambda x: np.max(x[rows]), axis=1).head()
```

```python
#The beauty of the applied function is that it allows flexibility in doing whatever manipulation that you desire, as the function you pass into apply can be 
# any customized function that you want. So let's say we wanted to divide the states into four categories; Northeast, Midwest, South, and West. 
# We can write customized function that returns this region based on the state, and the state regions information, maybe we looked up on Wikipedia.

def get_state_region(x):
    northeast=['Connecticut','Maine', 'Massachusetts', 'New Hampshire',
              'Rhode Island', 'Vermont', 'New York', 'New Jersey', 'Pennsylvania']
    midwest=['Illinois', 'Indiana', 'Michigan', 'Ohio', 'Wisconsin', 'Iowa', 
            'Kansas', 'Minnesota', 'Missouri''Kansas', 'Minnesota', 'Missouri']
    south= ['Delaware', 'Florida','Georgia', 'Maryland','North Carolina','South Carolina',
            'Virginia', 'district of Columbia', 'West Virginia','Alabama', 'Kentucky','Mississippi',
            'Tennessee', 'Arkansas', 'Louisiana' ,'Oklahoma', 'Texas']
    west=['Arizona','Colorado', 'Idaho', 'Montana', 'New Mexico','Nevada',
          'Utah','Wyoming','Alaska','California', 
          'hawaii' ,'Oregon' ,'Washington']
    
    if x in northeast:
        return 'Northeast'
    elif x in midwest:
        return 'Midwest'
    elif x in south:
        return 'South'
    else:
        return 'west'
```

```python
# So now we have a customized function. Let's say we want to create a new column called region, which shows the state region, we can use the customized function and 
# the apply function to do so. The customized function is supposed to work on the state Name column, STNAME. So we'll set the apply function to the state name column, 
# and pass the customized function into the Apply.

df['state_region'] = df['STNAME'].apply(lambda x: get_state_region(x))
```
